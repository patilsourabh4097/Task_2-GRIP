## Linear Regression

### Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Load datasets

```python
data = pd.read_csv("https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv")
```

```python
data.head()
```

### Visualizing dataset

```python
fig, ax = plt.subplots()

ax.scatter(data.Hours, data.Scores, c = data.Scores)
ax.set_xlabel('No. of hours studied')
ax.set_ylabel(' Score obtained')
ax.set_title(' No. of Hours studied Vs Scores')
```

```python
sns.regplot(x = 'Hours', y = 'Scores', data = data)
plt.xlabel('No. of Hours')
plt.ylabel('Scores')
plt.title('No. of Hours studied Vs Scores')
```

### Dividing dataset

```python
x=data.iloc[:, :-1].values
y=data.iloc[:, 1].values
```

### Train-test split

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train,pred = train_test_split(x,y, test_size = 0.25, random_state = 0)
```

### Model Selection - Linear Regression

As from visualizations, we can see that, dataset is having linear relationship. So, we use Linear Regression algorithm

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(x_train, y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
y_pred = model.predict(x_test)
```

```python
df = pd.DataFrame({'Actual':pred, 'Predicted':y_pred})
```

```python
df.head()
```

### Accuracy of the Model

```python
from sklearn.metrics import r2_score

r2_score(pred, y_pred)
```

### Checking prediction for 9.25 hours

```python
hr = [[9.25]]

score = model.predict(hr)
print('No. of hours studied {}'.format(hr[0]))
print('Score obtained {}'.format(score[0]))
```
